```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib 
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.drop(['Name','Ticket','Cabin'], axis=1)
df
```

```python
# å„ç‰¹å¾´é‡ã®æ¬ æå€¤ã‚’ã‚«ã‚¦ãƒ³ãƒˆ
df.isnull().sum()
```

```python
df = df.interpolate(method='linear')
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
Embarked_mapping = {'C': 0, 'Q': 1, 'S': 2}

df['Embarked'] = df['Embarked'].map(Embarked_mapping)

df
```

```python
SEX_mapping = {'male': 0, 'female': 1}

df['Sex'] = df['Sex'].map(SEX_mapping)
df
```

```python
df_target = pd.DataFrame(df, columns=['Survived'])
```

```python
df_target
```

```python
df.corr()
```

```python
df = df.drop(['Survived'], axis=1)
df
```

```python
df2 = df.drop(['PassengerId'], axis=1)
df2
```

```python
X= np.array(df2)
X.shape
```

```python
y= np.array(df_target)
y=y.reshape(-1)
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)
```

```python
X_train.shape
```

```python
# æ±ºå®šæœ¨ã«ã‚ˆã‚‹å­¦ç¿’
from sklearn.tree import DecisionTreeClassifier
tree_1 = DecisionTreeClassifier(random_state=0, max_depth=4,)
tree_1.fit(X_train, y_train)
```

```python
print('train: {:.3f}'.format(tree_1.score(X_train, y_train)))
print('test : {:.3f}'.format(tree_1.score(X_test, y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
tree_2 = RandomForestClassifier(random_state=0, n_estimators=10)
tree_2.fit(X_train, y_train)
```

```python
print('train: {:.3f}'.format(tree_2.score(X_train, y_train)))
print('test : {:.3f}'.format(tree_2.score(X_test, y_test)))
```

```python
index = 1
print('answer : %d' % y_test[index])
print('predict: %d' % tree_1.predict(X_test[index].reshape(1, -1)))
```

```python
param_grid = {'max_depth': [2, 3, 4, 5, 6],
              'learning_rate': [0.001, 0.01, 0.1, 1, 10]}
```

```python
from sklearn.model_selection import StratifiedKFold
kf_5 = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
gs_tree = GridSearchCV(GradientBoostingClassifier(), param_grid, cv=kf_5) 
```

```python
gs_tree.fit(X_train, y_train)
```

```python
gs_tree.best_params_
```

```python
from sklearn.ensemble import GradientBoostingClassifier
tree_3 = GradientBoostingClassifier(random_state=0, max_depth=5, learning_rate=0.1)
tree_3.fit(X_train, y_train)
```

```python
print('train accuracy: %.3f' % tree_3.score(X_train, y_train))
print('test  accuracy: %.3f' % tree_3.score(X_test, y_test))
```
